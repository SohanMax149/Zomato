```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context("talk")
import datetime
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
zomato['Cuisines'].fillna(zomato['Cuisines'].mode()[0], inplace=True)
```

```python
zomato.isnull().sum()
```

```python
cuisine = zomato.groupby(['Cuisines'], as_index=False)['Restaurant ID'].count()
cuisine.columns = ['Cuisines','Number of Resturants']
```

```python
cuisine_10 = (cuisine.sort_values(['Number of Resturants'],ascending=False)).head(10)
print(cuisine_10)
```

# EDA ON ZOMATO

```python
plt.figure(figsize=(14,8))
sns.barplot(cuisine_10['Cuisines'], cuisine_10['Number of Resturants'])
plt.xlabel('Cuisines', fontsize=20)
plt.ylabel('Number of Resturants', fontsize=20)
plt.title('Top 10 cuisines on Zomato', fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
Restaurant = zomato.groupby(['Restaurant Name'], as_index=False)['Restaurant ID'].count()
Restaurant.columns = ['Restaurant Name','Votes']
Restaurant_10 = (Restaurant.sort_values(['Votes'],ascending=False)).head(10)
plt.figure(figsize=(14,8))
sns.barplot(Restaurant_10['Restaurant Name'], Restaurant_10['Votes'])
plt.xlabel('Restaurant Name', fontsize=20)
plt.ylabel('Votes', fontsize=20)
plt.title('Top 10 Restaurant on Zomato', fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
zomato['City'].value_counts()
```

```python
zomato_main_cities = zomato[(zomato['City'] == 'Mumbai') | (zomato['City'] == 'New Delhi') | (zomato['City'] == 'Bangalore') | (zomato['City'] == 'Hyderabad') | (zomato['City'] == 'Ahmedabad')]
```

```python
zomato_main_cities.head()
```

# EDA ON MAIN CITIES INDIA

```python
cuisine = zomato_main_cities.groupby(['Cuisines'], as_index=False)['Restaurant ID'].count()
cuisine.columns = ['Cuisines','Number of Resturants']
cuisine_10 = (cuisine.sort_values(['Number of Resturants'],ascending=False)).head(10)
plt.figure(figsize=(14,8))
sns.barplot(cuisine_10['Cuisines'], cuisine_10['Number of Resturants'])
plt.xlabel('Cuisines', fontsize=20)
plt.ylabel('Number of Resturants', fontsize=20)
plt.title('Top 10 cuisines on Zomato on Main Cities', fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
Restaurant = zomato_main_cities.groupby(['Restaurant Name'], as_index=False)['Restaurant ID'].count()
Restaurant.columns = ['Restaurant Name','Votes']
Restaurant_10 = (Restaurant.sort_values(['Votes'],ascending=False)).head(10)
plt.figure(figsize=(14,8))
sns.barplot(Restaurant_10['Restaurant Name'], Restaurant_10['Votes'])
plt.xlabel('Restaurant Name', fontsize=20)
plt.ylabel('Votes', fontsize=20)
plt.title('Top 10 Restaurant on Zomato on Main Cities', fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
zomato_main_cities['Restaurant Name'].value_counts().head(5).plot(kind='bar', figsize=(10,10))
plt.xlabel('Name of restaurant', fontsize = 18)
plt.ylabel('Amount of each restaurant', fontsize = 18)
plt.title('Top 5 Restaurants with most outlets in Main Cities', fontsize = 25)
plt.figure()
```

```python
zomato_main_cities.groupby('Has Table booking')['Aggregate rating'].mean().sort_values(ascending=False).plot(kind='bar', figsize = (10,8), title='Relation between Booking and Aggregate rating')
```

```python
zomato_main_cities.groupby('Has Online delivery')['Aggregate rating'].mean().plot(kind='bar', figsize = (10, 8))
plt.ylabel('Aggregate rating') 
plt.show();
```
